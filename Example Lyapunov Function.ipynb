# Example Lyapunov Function

This example is taken from the book *Applied Nonlinear Control* by Slotine and Li, 
chapter 3 just after Definition 3.7.

The Lyapunov function, $V\left( \mathbf{x} \right)$ is given by

$V\left( \mathbf{x} \right) =\, \frac{1}{2}MR^2 x_2^2 + MRg \left(1 - \cos x_1 \right)$

## Make the Lyapunov function

Let's first create the Lyapunov function

```python
def lyapunov(X1, X2, g, M, R):
    return 0.5*M*R*R*X2*X2 + M*R*(1 - np.cos(X1))
```

## Plot the surface

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure(figsize=(8,8), dpi=120)
ax = fig.gca(projection='3d')

M = 1
R = 1
g = 9.8


# Make the data
x1 = np.arange(-np.pi,np.pi, 0.005)
x2 = np.arange(-4,4, 0.005)
X1, X2 = np.meshgrid(x1, x2)
Vx = lyapunov(X1, X2, g, M, R)

# Plot the surface
surf = ax.plot_surface(X1, X2, Vx, cmap=cm.jet,
                       linewidth=0, antialiased=False)

# Label axes
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('V(x)')

# Show the plot
plt.show()

# Save the plot
fig.savefig('lyapunov-example.svg', format='svg', dpi=1200)
```

## Plot the contours

```python
# Make the contour plot
cont_fig = plt.figure(figsize=(8,8), dpi=120)
axc = cont_fig.gca()
CS = plt.contour(X1, X2, Vx)
plt.clabel(CS, inline=1, fontsize=14)
axc.set_xlabel('x1', fontsize=16)
axc.set_ylabel('x2', fontsize=16)
axc.tick_params(axis='both', which='major', labelsize=14)

# Show the plot
plt.show()

# Save the plot
cont_fig.savefig('lyapunov-contours-example.svg', format='svg', dpi=1200)
```
